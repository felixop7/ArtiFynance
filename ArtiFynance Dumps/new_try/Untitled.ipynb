```
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, LSTM
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```
df = pd.read_csv("stock_data.csv")
```

```
train_data = df.iloc[:800, 1:2].values
test_data = df.iloc[800:, 1:2].values
```

```
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_train_data = scaler.fit_transform(train_data)
```

```
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(scaled_train_data.shape[1], 1)))
model.add(LSTM(units=50))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')
```

```
model.fit(scaled_train_data, scaled_train_data, epochs=100, batch_size=32)
```

```
inputs = df.iloc[len(df) - len(test_data) - 60:, 1:2].values
inputs = scaler.transform(inputs)
```

```
x_test = []
for i in range(60, len(inputs)):
    x_test.append(inputs[i-60:i, 0])
x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
```

```
predicted_stock_prices = model.predict(x_test)
predicted_stock_prices = scaler.inverse_transform(predicted_stock_prices)
```
